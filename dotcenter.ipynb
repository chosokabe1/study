```python
inpath = 'C:\ex\sen\data\max_square\\all'
outpath = 'C:\ex\sen\data\max_square\\all_visuable' 
labelpath = 'C:\ex\sen\data\max_square\canny\label'
```

```python
import os
import glob
import cv2
```

```python
for file in glob.glob(os.path.join(inpath,'*')):
  img = cv2.imread(file)
  with open(os.path.join(labelpath,os.path.basename(file).split('.')[0] + '.txt')) as f:
    s = f.read()
    x = s.split(' ')[0]
    y = s.split(' ')[1]
  
  img_x,img_y,_=img.shape
  x = int(float(x) * img_x)
  y = int(float(y) * img_y)
  cv2.circle(img, (x, y), 2, (255, 0, 0), thickness=-1)
  if x-112 < 0:
    pt1_x = 0
    pt2_x = 224
  elif x+112 > img_x - 1:
    pt1_x = img_x -1 -224
    pt2_x = img_x -1
  else:
    pt1_x = x - 112
    pt2_x = x + 112
  if y-112 < 0:
    pt1_y = 0
    pt2_y = 223
  elif y+112 > img_y - 1:
    pt1_y = img_y -1 -224
    pt2_y = img_y -1
  else:
    pt1_y = y - 112
    pt2_y = y + 112
  cv2.rectangle(img,pt1=(pt1_x,pt1_y),pt2=(pt2_x,pt2_y),color=(0,0,255),thickness=1)
  cv2.imwrite(os.path.join(outpath,os.path.basename(file)), img)
  print(os.path.join(outpath,os.path.basename(file)))
```
