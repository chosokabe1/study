```python
path = "egg_no_count.csv"
outpath = "egg_no_hist.csv"

import csv
count_0 = 0
count_1 = 0
count_2 = 0
count_3 = 0
count_4 = 0
count_5 = 0
count_6 = 0

with open(path) as f:
  with open(outpath, mode="w", newline="") as g:
    writer = csv.writer(g)
    writer.writerow(['no'])
    reader = csv.reader(f)
    for row in reader:
      print(row)
      if len(row) == 0:
        continue
      elif row[0] == "egg":
        writer.writerow(['0', str(count_0)])
        writer.writerow(['1', str(count_1)])
        writer.writerow(['2', str(count_2)])
        writer.writerow(['3', str(count_3)])
        writer.writerow(['4', str(count_4)])
        writer.writerow(['5', str(count_5)])
        writer.writerow(['6', str(count_6)])
        count_0 = 0
        count_1 = 0
        count_2 = 0
        count_3 = 0
        count_4 = 0
        count_5 = 0
        count_6 = 0
      
      if not len(row) == 3:
        continue
      elif row[2] == "0":
        count_0 += 1
      elif row[2] == "1":
        count_1 += 1
      elif row[2] == "2":
        count_2 += 1
      elif row[2] == "3":
        count_3 += 1
      elif row[2] == "4":
        count_4 += 1
      elif row[2] == "5":
        count_5 += 1
      elif row[2] == "6":
        count_6 += 1
      
    writer.writerow(['egg'])
    writer.writerow(['0', str(count_0)])
    writer.writerow(['1', str(count_1)])
    writer.writerow(['2', str(count_2)])
    writer.writerow(['3', str(count_3)])
    writer.writerow(['4', str(count_4)])
    writer.writerow(['5', str(count_5)])
    writer.writerow(['6', str(count_6)])
```
