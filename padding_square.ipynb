```python
def makedir(dir):
  if not os.path.exists(dir):
    os.makedirs(dir)
```

```python
import cv2
import glob

class Img:
  def __init__(self, path):
    im = cv2.imread(path)
    h, w, c = im.shape
    self.height = h
    self.width = w
    self.channel = c
```

```python
import numpy as np
import os

def padding_0(path, out_path):
  img = cv2.imread(path)
  h, w, _ = img.shape
  flag = 0
  if h < w:
    min_size = w
    newimg = np.zeros((min_size, w, 3))
    start = int((min_size - h) / 2)
    fin = int((min_size + h) / 2)
    newimg[start:fin, :] = img
    flag = 1
  
  if w < h:
    min_size = h
    newimg = np.zeros((h, min_size, 3))
    start = int((min_size - w) / 2)
    fin = int((min_size + w) / 2)
    newimg[:, start:fin] = img
    flag = 1

  if flag == 1:
    file_name = os.path.splitext(os.path.basename(path))[0]
    cv2.imwrite(os.path.join(out_path, file_name + '.jpg'), newimg)
  else:
    file_name = os.path.splitext(os.path.basename(path))[0]
    cv2.imwrite(os.path.join(out_path, file_name + '.jpg'), img)
```

```python
def padding_0_ioimg(img):
  h, w, _ = img.shape
  flag = 0
  if h < w:
    min_size = w
    newimg = np.zeros((min_size, w, 3))
    start = int((min_size - h) / 2)
    fin = int((min_size + h) / 2)
    newimg[start:fin, :] = img
    flag = 1
  
  if w < h:
    min_size = h
    newimg = np.zeros((h, min_size, 3))
    start = int((min_size - w) / 2)
    fin = int((min_size + w) / 2)
    newimg[:, start:fin] = img
    flag = 1

  if flag == 1:
    return newimg
  else:
    return img
```

```python
def resize_folder(inpath,outpath):
  for file in glob.glob(inpath + '/*'):
    img = cv2.imread(file)
    dst = cv2.resize(img, dsize=(50,50))
    cv2.imwrite(os.path.join(outpath,os.path.basename(inpath)),dst)
```
