```python
import glob
import os
import csv
from finetuning.classification_module import slide_is_it_egg

csv_header = ['full_img', 'slide', 'is_egg']
csv_body = []
for file_path in glob.glob(os.path.join(egg_slide_path,'*')):
  print(os.path.splitext(os.path.basename(file_path))[0][:-2])
  full_img_name = os.path.splitext(os.path.basename(file_path))[0][:-2]
  egg_slide_number = os.path.splitext(os.path.basename(file_path))[0][-1]
  is_egg = slide_is_it_egg.main(file_path)
  csv_body.append([full_img_name,egg_slide_number, is_egg])

with open('C:\ex\sen\data\padding_600min\cross\cross1_egg.csv', 'w', newline="") as f:
  writer = csv.writer(f)
  writer.writerow(csv_header)
  writer.writerows(csv_body)

csv_header = ['full_img', 'slide', 'is_egg']
csv_body = []
for file_path in glob.glob(os.path.join(no_slide_path,'*')):
  print(os.path.splitext(os.path.basename(file_path))[0][:-2])
  full_img_name = os.path.splitext(os.path.basename(file_path))[0][:-2]
  egg_slide_number = os.path.splitext(os.path.basename(file_path))[0][-1]
  is_egg = slide_is_it_egg.main(file_path)
  csv_body.append([full_img_name,egg_slide_number, is_egg])

with open('C:\ex\sen\data\padding_600min\cross\cross1_indis.csv', 'w', newline="") as f:
  writer = csv.writer(f)
  writer.writerow(csv_header)
  writer.writerows(csv_body)
```
