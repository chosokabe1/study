```python
no = "C:\ex\sen\data\padding_600min\cross\cross5_no.csv"
no_count_file = "C:\ex\sen\data\padding_600min\cross\cross5_no_count.csv"
egg = "C:\ex\sen\data\padding_600min\cross\cross5_egg.csv"
egg_count_file = "C:\ex\sen\data\padding_600min\cross\cross5_egg_count.csv"
import csv
```

```python
with open(egg_count_file, mode="w", newline="") as g:
  writer = csv.writer(g)
  writer.writerow(["full_img_name", "no_count", "egg_count"])
  with open(egg) as f:
    reader = csv.reader(f)
    no_count = 0
    egg_count = 0
    full_img_name = "shoki"
    for index,row in enumerate(reader):
      if index == 0:
        continue

      if full_img_name != 'img' + row[0] + row[1]:
        writer.writerow([full_img_name, no_count, egg_count])
        print(f"no_count = {no_count}, egg_count = {egg_count}")
        no_count = 0
        egg_count = 0
        full_img_name = 'img' + row[0] + row[1]

      print(row)
      if row[3] == "True":
        egg_count += 1 
      else:
        no_count += 1
```
