```python
import glob
import os
```

```python
inpath = "C:\ex\sen\data\max_square\canny\\annotation"
outpath = "C:\ex\sen\data\max_square\canny\label"
```

```python
for file in glob.glob(os.path.join(inpath,'*')):
  print(file)
  outfile = outpath + "\\" + os.path.basename(file)
  print(outfile)
  with open(file) as f:
    s = f.read()
    x = s.split(' ')[1]
    y = s.split(' ')[2]
  
  with open(outfile, mode="w") as f:
    f.write(x + ' ' + y)

```
